```python
import tensorflow as tf
print(tf.__version__)
```

scalar

```python
a = tf.constant(5)
b = tf.constant(10)

c = tf.add(a, b)
print(c)

d = tf.subtract(a, b)
print(d)

e = tf.multiply(a, b)
print(e)

f = tf.divide(a, b)
print(f)

g = tf.math.floordiv(a, b)
print(g)

p = tf.pow(b, 2)
print(p)
```

1D vector

```python
v = tf.constant([1,2,3,4])
u= tf.constant([5,6,7,8])
print(v)
print(u)

print("Add :",tf.add(v,u))
print("subtract : ",tf.subtract(v,u))
print("multiply : ",tf.multiply(v,u))
print("divide : ",tf.divide(v,u))
print("floordiv : ",tf.math.floordiv(v,u))
print("power : ",tf.pow(v,u))
```

2D

```python
m = tf.constant([[1, 2],
[3, 4]])
n = tf.constant([[5, 6],
[7, 8]])
print(m)
print(n)

print("Add :",tf.add(m,n))
print("subtract : ",tf.subtract(m,n))
print("multiply : ",tf.multiply(m,n))
print("divide : ",tf.divide(m,n))
print("floordiv : ",tf.math.floordiv(m,n))
print("power : ",tf.pow(m,n))
```
